```python
import json
import os
import numpy as np
# with open("/home/sonlt/workspace/markdocdown/EGGM/data/acc_cls_0,acc_cls_1,acc_cls_2 VS step_chart_data (2).json", "r") as f:
#     data = json.load(f)
with open("data/cos_cls_0,cos_cls_1,cos_cls_2 VS step_chart_data (1).json", "r") as f:
    data = json.load(f)
    
```

```python
len(data[0]['y'])
```

```python
a, t, v = data[0]['y'], data[1]['y'], data[2]['y']
a, t, v = np.array(a), np.array(t), np.array(v)
```

```python
a = a.reshape(20, -1).mean(axis=1)
t = t.reshape(20, -1).mean(axis=1)
v = v.reshape(20, -1).mean(axis=1)
```

```python
a.shape
```

```python
import matplotlib.pyplot as plt
import matplotlib

fig, ax = plt.subplots()
ax.plot(a, label='Audio', color='blue')
ax.plot(t, label='Text', color='orange')
ax.plot(v, label='Visual', color='green')

ax.legend(fontsize=14)
ax.set_ylabel('Magnitude', fontsize=16)
ax.set_xlabel('Epoch', fontsize=16)
xticks = np.arange(0, len(a), 2)
ax.set_xticks(xticks)
# x tick font size
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=14)
plt.savefig("figure/coeff_mosei.pdf", bbox_inches='tight', dpi=300)
plt.show()
```
